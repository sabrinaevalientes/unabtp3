```python
#Ejercicio 2:

class Punto:
    def __init__ (self, x, y):
        self.x = x
        self.y = y

    def eje_x(self):
      return self.x

    def eje_y(self):
      return self.y

    def impresion(self):
      return f"Punto(x={self.x}, y={self.y})"

    def opuesto(self):
      return Punto(-self.x, -self.y)

p1= Punto(3, 5)
print(p1.impresion())
p1_opuesto = (p1.opuesto())
print(p1_opuesto.impresion())
```
